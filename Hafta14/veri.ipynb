```python
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    else:
        return 'Fall'

df['season'] = df['arrival_date_month'].apply(get_season)

sns.barplot(x='season', y='is_canceled', data=df)
plt.title('Mevsimlere Göre İptal Oranı')
```

```python
# Ay sıralamasını takvimsel düzene sokmak için (isteğe bağlı)
df['arrival_date_month'] = pd.Categorical(
    df['arrival_date_month'],
    categories=['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December'],
    ordered=True
)

# Grup bazlı hesaplama
monthly_stats = df.groupby('arrival_date_month').agg(
    average_lead_time=('lead_time', 'mean'),
    cancel_count=('is_canceled', lambda x: (x == 1).sum())
)

print(monthly_stats)
```

```python
import matplotlib.pyplot as plt

# Veriyi yeniden hazırlayalım (güncel haliyle)
monthly_stats = df.groupby('arrival_date_month').agg(
    average_lead_time=('lead_time', 'mean'),
    cancel_count=('is_canceled', lambda x: (x == 1).sum())
).sort_index()  # Ayları takvim sırasına göre getirir (önceden Categorical yaptıysan gerek yok)

# Grafik
fig, ax1 = plt.subplots(figsize=(10, 6))

# Ortalama lead_time çubuğu (mavi)
ax1.bar(monthly_stats.index, monthly_stats['average_lead_time'], color='skyblue', label='Ortalama Lead Time')
ax1.set_ylabel('Ortalama Lead Time (gün)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# İkinci y ekseni: iptal eden kişi sayısı (kırmızı çizgi)
ax2 = ax1.twinx()
ax2.plot(monthly_stats.index, monthly_stats['cancel_count'], color='red', marker='o', label='İptal Sayısı')
ax2.set_ylabel('İptal Eden Sayısı', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Başlık ve düzenleme
plt.title('Her Ay için Ortalama Lead Time ve İptal Eden Sayısı')
fig.tight_layout()
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Göster
plt.show()
```

```python
# Ay isimlerini sayısal sıralama için kategorik ve sıralı yapalım
aylar = ['January', 'February', 'March', 'April', 'May', 'June', 
         'July', 'August', 'September', 'October', 'November', 'December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=aylar, ordered=True)

# Yıllar ve aylara göre grup oluşturup istatistikleri alalım
monthly_stats = df.groupby(['arrival_date_year', 'arrival_date_month']).agg({
    'lead_time': 'mean',
    'is_canceled': 'mean',  # iptal oranı
    'hotel': 'count'        # rezervasyon sayısı
}).rename(columns={'hotel': 'reservation_count'}).reset_index()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_stats, x='arrival_date_month', y='is_canceled', hue='arrival_date_year', marker='o')
plt.title('Yıllara Göre Aylık İptal Oranları')
plt.ylabel('İptal Oranı')
plt.xlabel('Ay')
plt.legend(title='Yıl')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_stats, x='arrival_date_month', y='lead_time', hue='arrival_date_year', marker='o')
plt.title('Yıllara Göre Aylık Ortalama Lead Time')
plt.ylabel('Ortalama Lead Time (Gün)')
plt.xlabel('Ay')
plt.legend(title='Yıl')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_stats, x='arrival_date_month', y='reservation_count', hue='arrival_date_year', marker='o')
plt.title('Yıllara Göre Aylık Rezervasyon Sayısı')
plt.ylabel('Rezervasyon Sayısı')
plt.xlabel('Ay')
plt.legend(title='Yıl')
plt.xticks(rotation=45)
plt.show()
```
