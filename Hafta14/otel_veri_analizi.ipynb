```python
# Lead time'ı gruplara ayıralım
bins = [0, 30, 60, 90, 120, 150, 180, 210, 300, 400, 600, df['lead_time'].max()]
labels = ['0-30', '31-60', '61-90', '91-120', '121-150', '151-180', '181-210', '211-300', '301-400', '401-600', '600+']

df['lead_time_group'] = pd.cut(df['lead_time'], bins=bins, labels=labels)

# Gruplara göre iptal oranı
iptal_orani_grup = df.groupby('lead_time_group')['is_canceled'].mean().reset_index()

# Grafik çizimi
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(x='lead_time_group', y='is_canceled', data=iptal_orani_grup)
plt.xlabel('Lead Time (Gün Aralığı)')
plt.ylabel('İptal Oranı')
plt.title('Lead Time Gruplarına Göre İptal Oranları')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bu daha anlamlı oldu gün yaklaştıkça iptal etme oranları daha az 
```

```python
fixed_bins = [0, 31, 61, 91, 121, 151, 181, 211, 301, 401, 601]

# df['lead_time'].max() değerini alalım
max_lead_time_in_data = df['lead_time'].max()

# Bins listesini oluştururken, max_lead_time_in_data'nın fixed_bins'in son elemanından
# büyük olup olmadığını kontrol edelim. Eğer büyükse, onu da ekleyelim.
# Ayrıca, pd.cut'ın sağ tarafı hariç tutması için +1 ekliyoruz.
if max_lead_time_in_data >= fixed_bins[-1]:
    bins = fixed_bins + [max_lead_time_in_data + 1]
else:
    # Eğer max_lead_time_in_data fixed_bins'in son elemanından küçükse,
    # sadece max_lead_time_in_data'yı kapsayan fixed_bins kısmını alalım.
    bins = [b for b in fixed_bins if b <= max_lead_time_in_data + 1]
    if max_lead_time_in_data + 1 > bins[-1]: # Eğer son bin değeri max_lead_time'ı kapsamıyorsa ekle
        bins.append(max_lead_time_in_data + 1)

# Bins listesinin kesinlikle artan ve benzersiz olduğundan emin olalım
bins = sorted(list(set(bins)))


# Etiketler
# Etiketleri bins listesine göre dinamik olarak oluşturalım
labels = []
for i in range(len(bins) - 1):
    lower_bound = bins[i]
    upper_bound = bins[i+1] - 1 # pd.cut right=False olduğu için
    if i == len(bins) - 2 and upper_bound >= 600: # Son aralık için özel etiket
        labels.append(f'{lower_bound}+')
    else:
        labels.append(f'{lower_bound}-{upper_bound}')

# Eğer labels sayısı bins'in bir eksiği kadar değilse veya son etiket sorunluysa düzeltme
# Bu kısım, bins listesinin dinamik yapısına göre labels'ı tam olarak eşleştirmek için
# biraz daha karmaşık olabilir. En basit yol, manuel olarak verdiğiniz labels'ı kullanmaktır,
# ancak bins listesi değiştiğinde labels'ın da uyumlu olması gerekir.
# En güvenli yol, labels'ı da dinamik olarak oluşturmaktır.
# Ancak, sizin belirttiğiniz sabit labels listesini korumak istiyorsanız,
# bins listesinin bu labels'a tam olarak uyduğundan emin olmalıyız.
# Bu durumda, bins listesini labels'a göre ayarlamak daha mantıklı.

# Sizin istediğiniz sabit etiketler:
labels = ['0-30', '31-60', '61-90', '91-120', '121-150', '151-180', '181-210', '211-300', '301-400', '401-600', '600+']

# Bins listesini labels'a uygun hale getirelim ve maksimum lead_time'ı kapsayalım
# Bu, labels'a tam olarak uyan bir bins listesi oluşturur ve son binin max_lead_time'ı kapsamasını sağlar.
bins = [0, 31, 61, 91, 121, 151, 181, 211, 301, 401, 601]
if max_lead_time_in_data >= 600: # Eğer veride 600+ lead_time varsa
    bins.append(max_lead_time_in_data + 1)
else: # Eğer yoksa, 601'den sonraki binleri atmalıyız ki labels'a uysun
    # labels'ın son elemanına karşılık gelen bin'i bulalım
    # Bu, labels'ın uzunluğuna göre bins'i dinamik olarak keser.
    # Örneğin, max_lead_time 350 ise, 401-600 ve 600+ aralıklarına ihtiyacımız yok.
    # Bu durumda bins ve labels'ı kısaltmalıyız.
    # Bu senaryo için daha dinamik bir labels ve bins oluşturma yaklaşımı daha iyidir.
    # Ancak, sizin sabit labels isteğiniz olduğu için, bins'i labels'a göre ayarlayalım.
    # Eğer max_lead_time 600'den küçükse, 601'den sonraki binleri atmalıyız.
    bins = [b for b in bins if b <= max_lead_time_in_data + 1]
    # Eğer son bin değeri max_lead_time'ı kapsamıyorsa ekle
    if max_lead_time_in_data + 1 > bins[-1]:
        bins.append(max_lead_time_in_data + 1)

# labels listesini bins listesinin uzunluğuna göre kırpalım
labels = labels[:len(bins)-1]


# pd.cut kullanarak lead_time_group sütununu oluşturma
# right=False: Sol taraf dahil, sağ taraf hariç (örn: [0, 31) -> 0'dan 30'a kadar)
# include_lowest=True: En düşük değeri (0) ilk aralığa dahil et
df['lead_time_group'] = pd.cut(df['lead_time'], bins=bins, labels=labels, right=False, include_lowest=True)

# Gruplara göre iptal oranını hesaplama
cancel_rate_by_group = df.groupby('lead_time_group')['is_canceled'].mean().reset_index()
cancel_rate_by_group['Cancel_Rate (%)'] = cancel_rate_by_group['is_canceled'] * 100

# NaN değerleri olan grupları düşürelim (eğer bir grupta hiç veri yoksa oluşabilir)
cancel_rate_by_group.dropna(subset=['Cancel_Rate (%)'], inplace=True)

# Görselleştirme
plt.figure(figsize=(14, 7))
sns.barplot(x='lead_time_group', y='Cancel_Rate (%)', data=cancel_rate_by_group,
            palette='coolwarm', hue='lead_time_group', legend=False) # FutureWarning'ı çözmek için hue ve legend=False eklendi
plt.title('Lead Time Gruplarına Göre Ortalama İptal Oranı') # Title of the plot
plt.xlabel('Lead Time Aralığı (Gün)') # X-axis label
plt.ylabel('İptal Oranı (%)') # Y-axis label
plt.xticks(rotation=45, ha='right') # Etiketlerin okunabilirliği için döndürme
plt.tight_layout() # Adjust plot to ensure everything fits
plt.show()
```
