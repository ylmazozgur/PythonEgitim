```python
########
### IQR yöntemi ile outlier tespiti
########
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
# fare sütunu için outlier tespiti
fare_outliers = detect_outliers_iqr(df, 'fare')
print(f"fare outlier sayısı: {len(fare_outliers)}") #Aykırı değerleri içeren satırlardan oluşan bir DataFrame

# Aykırı değerleri tespit edip:
# Dilersek temizleyebiliriz (drop)
# Dilersek analizden hariç tutabiliriz
# Ya da ayrı analiz yapabiliriz (yüksek ücret ödeyenler kimler?)
```

```python
########
### Özel Aggregation Fonksiyonları
########
def custom_stats(series):
    stats = {
        'count': series.count(),
        'mean': series.mean(),
        'std': series.std(),
        'min': series.min(),
        'max': series.max(),
        'range': series.max() - series.min(),
        'skewness': series.skew()
    }
    return pd.Series(stats)

#Yaş dağılımının asimetrisi (çarpıklık)
#her sınıf için yaşın detaylı istatistikleri elde ederiz.
detailed_stats = df.groupby('pclass')['age'].apply(custom_stats)
```

```python
### Rolling window analizi (yaş sıralaması için)
# Zamanla değişen trendleri, dalgalanmaları ve istatistikleri izlemeyi sağlar.
# Ani artış/düşüşleri daha net görmenizi sağlar.
# Finans, hava durumu, sensör verileri gibi zaman serilerinde çok yaygın kullanılır.

df_sorted = df.sort_values('age').reset_index(drop=True)
df_sorted['Age_rolling_mean'] = df_sorted['age'].rolling(window=10, min_periods=1).mean()
# window -> Kaç satırlık veri alınacağı bilgisi
# min_periods -> Minimum veri sayısı (eksik varsa ne yapılacak)
```

```python
# Memory optimization
def optimize_memory(df):
    start_mem = df.memory_usage(deep=True).sum() / 1024**2
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object and not pd.api.types.is_categorical_dtype(col_type):
            c_min = df[col].min()
            c_max = df[col].max()
            
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                    
            else:  # float
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
    
    end_mem = df.memory_usage(deep=True).sum() / 1024**2
    print(f'Memory usage decreased from {start_mem:.2f} MB to {end_mem:.2f} MB '
          f'({100 * (start_mem - end_mem) / start_mem:.1f}% reduction)')
    
    return df
```
