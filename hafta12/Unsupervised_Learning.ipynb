# Denetimsiz Öğrenme (Unsupervised Learning)

Denetimsiz Öğrenme, **etiketlenmemiş verilerdeki gizli yapıları ve örüntüleri** keşfetmek için kullanılan bir makine öğrenmesi türüdür.

* **Denetimli Öğrenme (Supervised Learning)'**de algoritmaya hem girdiyi (örneğin bir kaplanın resmi) hem de çıktıyı (resmin "kaplan" etiketi) veririz. Model, girdiler ve çıktılar arasındaki ilişkiyi öğrenir.
* **Denetimsiz Öğrenme'**de ise, algoritmaya sadece girdiyi (örneğin binlerce haycan resmi) veririz, ancak bu resimlerin ne olduğunu söylemeyiz ("kedi", "kaplan", "fare" gibi etiketler yoktur). Algoritmanın görevi, bu verileri kendi başına analiz ederek içindeki doğal gruplamaları, anormallikleri veya ilişkileri bulmaktır.

Denetimsiz öğrenmenin temel amacı **tahmin yapmak değil, veriyi anlamak ve tanımlamaktır.**

<br></br>

* **Denetimsiz Yaklaşım:** Legoları renge, şekle veya boyuta göre gruplara ayırmaya başlarız. Kırmızıları bir yığına, mavileri başka bir yığına toplarız. Bu, verinin iç yapısını keşfetmektir.

## 1. Kümeleme (Clustering)

Amacı, veri noktalarını birbirine benzer olanları aynı gruba (küme), benzemeyenleri ise farklı gruplara ayırmaktır. Bir kümedeki veri noktaları birbirine çok benzerken, farklı kümelerdeki veri noktaları birbirinden oldukça farklıdır.

### 1.1 K-Means (K-Ortalamalar):

Genellikle, verilerdeki benzer özelliklere sahip öğeleri gruplamak amacıyla kullanılır. K-Means, genellikle daha büyük veri kümelerindeki desenleri tanımlamak için tercih edilir.

**K-Means Algoritması Adımları:**
1. **K Değerini Belirleme:**

    İlk adımda, kaç tane kümeye (cluster) ayıracağımıza karar vermemiz gerekiyor. Bu "K" sayısı, algoritmanın başlangıç parametresidir ve verilerin doğal yapısına göre seçilmelidir. Genellikle denemelerle veya bazı yöntemlerle (örneğin, **Elbow Method**, **Silhouette Analysis**) K sayısı belirlenir.

2. **Başlangıçta K Merkezinin Seçilmesi:**

    K-means algoritması, her bir küme için bir merkez (centroid) seçer. Başlangıçta, bu merkezler rastgele seçilir. Bu merkezler, her bir kümenin temsilcisi olacak ve tüm veri noktaları bu merkezlere göre gruplandırılacaktır.

3. **Veri Noktalarının Küme Merkezlerine Atanması:** 

    Her bir veri noktası, en yakın merkezle (centroid) ilişkilendirilir. Yani, her bir veri noktası, merkezlerin bulunduğu mesafeye göre en yakın olduğu kümeye atanır. Bu işlem, her veri noktasının hangi kümeye ait olduğunu belirler.

4. **Merkezlerin Güncellenmesi:**

    Veri noktaları her kümeye atandıktan sonra, her bir kümenin yeni merkezi hesaplanır. Küme merkeznin yeni değeri, o kümeye ait tüm veri noktalarının ortalaması (veya ağırlıklı ortalaması) alınarak belirlenir.

5. **Tekrar Etme:**

    Adımlar 3 ve 4, kümelerin merkezleri stabil olana kadar tekrarlanır. Yani, merkezler değişmeğini veya çok küçük bir değişim olduğu zaman algoritma durur. Bu durumda, her veri noktası son kez kümesine atanmış olur.

Yani uzun lafın kısası:
* Başlangıçta K merkez seçilir.
* Veri noktaları her kümeye atanır.
* Kümelerin merkezleri güncellenir.
* Bu işlem merkezler sabit kalana kadar devam eder.

**Küme Merkezi (Centroid) Hesaplama**

Bir kümenin merkezi, o kümedeki tüm veri noktalarının ortalaması alınarak hesaplanır. Eğer her bir veri noktası $x_i = (x_{i1},x_{i2}, ... ,x_{in})$ ve kümenin $n$ veri noktası varsa, küme merkeznin her bir özelliği (veya boyutu) şu şekilde hesaplanır:

$$
C_j = \frac {1}{n} \sum _{i=1}^{n} x_i
$$

* $C_j$: Küme merkezidir. Yani $j$-inci kümedeki tüm veri noktalarının ortalamasıdır.
* $n$: Veri noktalarının sayısıdır. Yani $C_j$ kümesi için kaç tane veri noktası olduğunu belirtir.
* $\sum _{i=1}^{n} x_i$ --> Tüm veri noktalarının toplamı 



**K-Means Algoritmasının Avantajları:**

* **Basit ve Etkili:** Hem anlayışı kolay hem de uygulaması basit olan bir algoritmadır.
* **Hızlı:** Çoğu durumda hızlı çalışır ve büyük veri setlerine uygulanabilir.
* **Düşük Hesaplama Maliyeti:** Algoritma, özellikle büyük veri kümesi üzerinde, hesaplama açısından verimli olabilir

**K-Means Algoritmasının Dezavantajları:**

* **K Değerinin Belirlenmesi:** K'nin nasıl belirleneceği her zaman net değildir. Genellikle deneysel bir süreçtir.
* **Başlangıç Merkezlerine Duyarlılık:** Kümelerin başlangıç merkezlerinin seçimi algoritmanın sonucunu etkileyebilir. Farklı başlangıçlar, farklı sonuçlar verebilir.
* **Yuvarlak Kümelere Yönelik:** K-Means, veri kümelerinin genellikle yuvarlak veya küresel olduğu varsayımıyla çalışır. Eğer kümeler lineer ya da farklı şekillerde ise, algoritma yeterince iyi sonuç veremeyebilir.
* **Aykırı Değerlere Duyarlılık:** Aykırı veriler (outliers), kümelerin merkezlerini bozarak yanlış kümeler oluşturulmasına neden olabilir.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler # Verileri Ölçeklendirme: K-Means, uzaklık temelli bir algoritma olduğu için özelliklerin farklı ölçeklerde olması (örneğin fare 0-500 arası iken pclass 1-3 arası) sonucu olumsuz etkiler. StandardScaler ile tüm özellikleri aynı ölçeğe getireceğiz.
from sklearn.cluster import KMeans

# Seaborn kütüphanesinden Titanic veri setini yükleyelim
df = sns.load_dataset('titanic')


# Kümeleme için kullanacağımız özellikleri seçelim
features = ['pclass', 'age', 'sibsp', 'parch', 'fare']
df_cluster = df[features].copy()

# 1. Eksik 'age' verilerini medyan ile doldurma
df_cluster.fillna({"age": df_cluster['age'].median()}, inplace=True)

# 2. Veri setinde hiç eksik veri kalıp kalmadığını kontrol etme
print("\nEksik Veri Kontrolü:")
print(df_cluster.isnull().sum())

# 3. Verileri ölçeklendirme
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cluster)

# Ölçeklenmiş veriyi bir DataFrame'e dönüştürelim (görmek için)
scaled_df = pd.DataFrame(scaled_features, columns=features)
print("\nÖlçeklendirilmiş Veri Örneği:")
print(scaled_df.head())
```

```python
# En önemli sorulardan biri şudur: "Veri setini kaç kümeye (K) ayırmalıyız?" Bunun için en yaygın yöntem Dirsek Metodu'dur.
# Bu metot, farklı K değerleri için K-Means modelini çalıştırır ve her K değeri için "inertia" (küme içi hata kareleri toplamı) değerini hesaplar. Inertia, küme merkezlerinin kendi kümelerindeki noktalara olan uzaklıklarının toplamıdır. Amacımız, inertia'yı düşüren ama gereksiz sayıda küme oluşturmayan bir K değeri bulmaktır.
inertia_values = []
k_range = range(1, 11) # 1'den 10'a kadar K değerlerini deneyeceğiz

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init uyarısını engellemek için
    kmeans.fit(scaled_features)
    inertia_values.append(kmeans.inertia_)

# Dirsek grafiğini çizelim
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, marker='o', linestyle='--')
plt.title('Dirsek Metodu (Elbow Method)')
plt.xlabel('Küme Sayısı (K)')
plt.ylabel('Inertia Değeri')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

**Grafik Yorumu:** Grafiğe baktığımızda, K=3 veya K=4 noktasında belirgin bir "dirsek" (kırılma) olduğunu görüyoruz. Bu noktalardan sonra inertia'daki düşüş yavaşlıyor. Bu, 3 veya 4 kümenin bu veri seti için mantıklı bir ayrım olduğunu gösterir. Örneğimiz için K=4 seçerek devam edelim.

```python
# K=4 için K-Means modelini oluşturalım ve eğitelim
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans.fit(scaled_features)

# Her bir yolcunun hangi kümeye ait olduğunu bulalım
cluster_labels = kmeans.labels_

# Bu küme etiketlerini orijinal veri setimize yeni bir sütun olarak ekleyelim
df['cluster'] = cluster_labels

# print("\nHer Yolcu İçin Atanan Küme Etiketleri:")
# print(df.head())



### Modelimiz yolcuları 4 farklı gruba ayırdı. Peki bu gruplar ne anlama geliyor? Şimdi her bir kümenin ortalama özelliklerine bakarak profillerini çıkaralım.

# Hayatta kalma bilgisini de analize dahil edelim
analysis_cols = ['survived', 'pclass', 'age', 'fare', 'cluster']

# Her kümenin ortalama değerlerini hesaplayalım
cluster_analysis = df[analysis_cols].groupby('cluster').mean()
print("\nKüme Profilleri (Ortalama Değerler):")
print(cluster_analysis)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Veri seti oluştur

# küme 1
x1 = np.random.normal(25,5,2000) # # Ortalama 25, standart sapması 5 olan, normal dağılıma uyan, 2000 değer üret
y1 = np.random.normal(25,5,2000)

# küme 2
x2 = np.random.normal(55,5,2000)
y2 = np.random.normal(60,5,2000)

# küme 3
x3 = np.random.normal(55,5,2000)
y3 = np.random.normal(15,5,2000)

x = np.concatenate((x1,x2,x3), axis = 0) # concatenate fonksiyonu, verilen dizileri belirli bir eksen boyunca birleştirir.
y = np.concatenate((y1,y2,y3), axis = 0)

dictionary ={"x":x, "y":y}

data = pd.DataFrame(dictionary)
data.head()

plt.figure()
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hiyerarşik Kümeleme Yöntemi için Oluşturulan Veri Seti")
plt.show()
```

### 1.2 Hiyerarşik Kümeleme (Hierarchical Clustering):

Veri noktalarını bir hiyerarşi veya bir "ağaç" yapısı (dendrogram) içinde gruplayan bir kümeleme analiz yöntemidir.

Bu algoritmanın temel amacı, benzer veri noktalarını aynı kümede, benzemeyenleri ise farklı kümelerde toplamaktır. "Hiyerarşik" olarak adlandırılmasının sebebi, bu kümelerin iç içe geçmiş bir yapıda oluşturulmasıdır.

Temel iki yaklaşım vardır:
1. **Birleştirici (Agglomerative) Hiyerarşik Kümeleme:** En yaygın kullanılan yaklaşımdır. "Aşağıdan yukarıya" (bottom-up) bir mantıkla çalışır.
2. **Bölücü (Divisive) Hiyerarşik Kümeleme:** "Yukarıdan aşağıya" (top-down) bir mantıkla çalışır.

**Agglomerative Hierarchical Clustering Adımları**

**Adım 1: Tüm Noktaları Ayrı Küme Olarak Kabul Et**
* Başlangıçta her veri noktası bir kümedir (n küme) .

**Adım 2: Benzerlik Matrisini Hesapla**
* Noktalar arasındaki mesafeyi (Öklid, Manhattan, Kosinüs vb.) hesapla.

**Adım 3: En Yakın İki Kümeyi Birleştir**
* **Bağlantı (Linkage) Kriterleri** ile küme çiftleri arasındaki benzerlik belirlenir:
  * **Single Linkage:** İki kümedeki en yakın noktaların mesafesi.
  * **Complete Linkage:** İki kümedeki en uzak noktaların mesafesi.
  * **Average Linkage:** Tüm nokta çiftlerinin ortalama mesafesi.
  * **Ward's Method:** Kümeler birleştikçe artan varyansı minimize eder (En yaygın yöntem).

**Adım 4: Yeni Kümeyi Matrise Ekle ve Güncelle**
* Birleştirilen kümeleri yeni bir küme olarak ele al, mesafe matrisini güncelle.

**Adım 5: Tek Küme Kalana Kadar Tekrarla**
* Tüm noktalar tek bir küme olana kadar Adım 3-4'ü tekrarla.

**Adım 6: Dendrogram Çiz ve Küme Sayısını Belirle**
* Dendrogramdaki **dikey mesafeler,** kümelerin birleşme sırasını ve benzerlik seviyesini gösterir.
* **Yatay bir çizgi çizerek** optimal küme sayısı seçilir.

**Avantajları:**
* Küme sayısı önceden bilinmek zorunda değil.
* **Yorumlanabilirlik:** Dendrogram ile veri yapısının hiyerarşisini sezgisel bir şekilde gösterir.
* **Esneklik:** Farklı bağlantı kriterleri (linkage) kullanılabilir.

**Dezavantajları:**
* **Hesaplama karmaşıklığı yüksek $O(n^3)$,** büyük veri setlerinde yavaş çalışır.
* **Geriye dönüşü yoktur.** Birleştirme veya bölme adımı bir kez yapıldığında geri alınamaz. Bu, başlangıçta yapılan bir kararın sonradan düzeltilemeyeceği anlamına gelir.
* K-Means'e göre daha az ölçeklenebilir.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Veri seti oluştur

# küme 1
x1 = np.random.normal(25,5,20) # # Ortalama 25, standart sapması 5 olan, normal dağılıma uyan, 20 değer üret
y1 = np.random.normal(25,5,20)

# küme 2
x2 = np.random.normal(55,5,20)
y2 = np.random.normal(60,5,20)

# küme 3
x3 = np.random.normal(55,5,20)
y3 = np.random.normal(15,5,20)

x = np.concatenate((x1,x2,x3), axis = 0) # concatenate fonksiyonu, verilen dizileri belirli bir eksen boyunca birleştirir.
y = np.concatenate((y1,y2,y3), axis = 0)

dictionary ={"x":x, "y":y}

data = pd.DataFrame(dictionary)
data.head()

plt.figure()
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hiyerarşik Kümeleme Yöntemi için Oluşturulan Veri Seti")
plt.show()
```

```python
# %% dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

merg = linkage(data, method="ward")
dendrogram(merg, leaf_rotation= 90)
plt.xlabel("Veri Noktaları")
plt.ylabel("Öklid Mesafesi")
plt.show()
```

### 1.3 DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

DBSCANS, "gürültü içeren uygulamalar için yoğunluk tabanlı mekansal kümeleme" anlamına gelir. Bu algoritma veri noktalarının **yoğun olduğu bölgeleri** küme olarak tanımlanır ve seyrek bölgelerdeki noktaları ise **"gürültü (noise)"** veya **"aykırı değer" (outlier)** olarak etiketler.
<br></br>

**Algoritma Mantığı**
1. **(epsilon - $\epsilon$)**: Bir veri noktasının "komşuluğunu" tanımlayan yarıçaptır. Bir noktanın epsilon mesafesindeki tüm noktalar, o noktanın komşusu olarak kabul edilir.
2. **min_samples (veya min_pts)**: Bir bölgenin "yoğun" olarak kabul edilmesi için epsilon yarıçapı içinde bulunması gereken minimum komşu sayısıdır.
3. **Nokta Türleri:** Bu iki parametreye (epsilon ve min_samples) göre, algoritma her bir veri noktasını üç kategoriden birine atar:
   * **Çekirdek Nokta (Core Point):** Bir noktanın epsilon yarıçapı içindeki komşu sayısı (kendisi dahil) min_samples'tan büyük veya eşitse, bu nokta bir **çekirdek noktadır.** Bu noktalar, bir kümenin kalbinde, yani en yoğun bölgesinde yer alır.
   * **Sınır Noktası (Border Point):** Bir çekirdek noktanın epsilon komşuluğu içinde olan, ancak kendisi bir çekirdek nokta olmayan (yani eps komşuluğunda min_samples'tan daha az komşusu olan) noktalardır. Bu noktalar, bir kümenin kenarında yer alır. Bir kümeye aittirler ama küme başlatamazlar.
   * **Gürültü Noktası (Noise Point / Outlier):** Ne bir çekirdek nokta ne de bir sınır noktası olan noktalardır. Yani, herhangi bir çekirdek noktanın epsilon komşuluğunda değildirler. Bu noktalar hiçbir kümeye ait değildir.

![](DBSCAN_1.png)

![](DBSCAN_2.png)

**Nasıl Çalışır?**
1. **Başlangıç:** Algoritma, rastgele bir veri noktası seçerek başlar.
2. **Komşuluk Kontrolü:** Seçilen noktanın eps yarıçapı içindeki komşularını sayar.
3. **Çekirdek/Gürültü Kararı:**
   * Eğer komşu sayısı min_samples'dan büyük veya eşitse, bu nokta bir **çekirdek nokta** olarak işaretlenir ve yeni bir küme oluşturulur.
   * Eğer komşu sayısı min_samples'dan az ise, bu nokta geçici olarak **gürültü** olarak etiketlenir. (Daha sonra başka bir çekirdek noktanın komşusu olarak bulunup bir kümeye dahil olabilir.)
4. **Kümeyi Genişletme:**
   * Eğer bir çekirdek nokta bulunduysa, onun eps komşuluğundaki tüm noktalar bu yeni kümeye eklenir.
   * Bu komşular arasında yeni **çekirdek noktalar** varsa, onların da komşuları sırayla kümeye dahil edilir.
   * Bu süreç, kümeye eklenecek yeni çekirdek nokta kalmayana kadar "yoğunlukla ulaşılabilir" (density-reachable) tüm noktaları kapsayacak şekilde devam eder.
5. **Tekrarlama:** Bir küme tamamen bulunduktan sonra, algoritma daha önce ziyaret edilmemiş başka bir veri noktasına geçer ve 2.adımdan itibaren süreci tekrarlar.
6. **Bitiş:** Tüm veri noktaları ziyaret edildiğinde (ya bir kümeye atanmış ya da gürültü olarak etiketlenmiş), algoritma sona erer.

**Avantajları:**
* **Küme Sayısı Gerekmez**
* **Keyfi Şekiller:** Farklı geometriklere sahip kümeleri bulabilir.
* **Gürültü Tespiti:** Anomali tespiti için doğal bir yeteneği vardır.
* **Parametre Anlamlılığı:** eps ve min_sample parametreleri, K-Means'in K değerine göre daha sezgisel ve alan bilgisiyle belirlenmesi daha kolaydır.
* **Sıralamaya Bağımlı Değil:** Veri noktalarının hangi sırayla işlendiği (birkaç sınır noktası durumu hariç) sonucu değiştirmez.

**Dezavantajları:**
* **Farklı Yoğunluktaki Kümelere Karşı Hassasiyet:** Eğer veri setinde hem çok yoğun hem de az yoğun kümeler bir aradaysa, tüm kümeleri doğru bir şekilde yakalayacak tek bir epis ve min_samples kombinasyonu bulmak zor olabilir.
* **Yüksek Boyutlu Verilerde Zorluk:** "Özelliklerin Laneti" (Curse of Dimensionality) nedeniyle, yüksek boyutlu uzaylarda mesafe kavramı anlamını yitirebilir. Bu durumda eps parametresini doğru seçmek çok zorlaşır.
* **Parametre Seçimi:** Her ne kadar sezgisel olsalar da, eps ve min_samples parametrelerinin doğru seçimi sonuçları büyük ölçüde etkiler ve bazen deneme-yanılma gerektirir. (Genellikle k-distance graph gibi yöntemlerle optimal eps değeri bulunmaya çalışılır.)

**Kullanım Alanları**
* **Anomali Tespiti:** Kredi kartı sahtekarlığı, ağ sızma girişimleri gibi normalden sapan davranışların tespiti.
* **Coğrafi Veri Analizi:** Harita üzerinde belirli yoğunluktaki yerleşim yerlerini, suç odaklarını veya hastalık kümelerini bulmak.
* **Görüntü Segmentasyonu:** Bir görüntüdeki benzer doku veya renge sahip bölgeleri gruplamak.
* **Biyoinformatik:** Benzer genetik özelliklere sahip protein ailelerini tanımlamak.
* **Müşteri Segmentasyonu:** Davranışları ortak bir "yoğunluk" göstermeyen, aykırı müşteri profillerini tespit etmek.

## 2. Boyut Azaltma (Dimensionality Reduction)

### 2.1 PCA (Principal Component Analysis / Temel Bileşenler Analizi):

PCA, bir veri setindeki **değişkeler arasındaki ilişkileri** analiz ederek, veriyi daha az boyutla ifade etmeyi sağlayan bir istatistiksel yöntemdir.
<br></br>

**Neden Boyut Azaltmaya İhtiyaç Duyarız?**
1. **Özelliklerin Laneti (Curse of Dimensionality):** Bir veri setindeki özellik (boyut) sayısı arttıkça, veri seyrelir ve anlamlı örüntüler bulmak için gereken veri miktarı katlanarak artar. Bu durum, modellerin performansını düşürür ve aşırı öğrenmeye (overfitting) yol açar.
2. **Veri Görselleştirme:** İnsan beyni 3 boyuttan sonrasını algılayamaz. 100 özellikli bir veriyi görselleştirmek imkansızdır. PCA, bu veriyi 2 veya 3 temel bileşene indirgeyerek görselleştirmemizi ver verinin yapısını anlamamızı sağlar.
3. **Gürültü Azaltama:** Genellikle daha az varyansa sahip olan bileşenler, veri setindeki gürültüyü temsil eder. PCA bu bileşenleri atarak veriyi temizlemeye yardımcı olabilir.
4. **Hesaplama Maliyetini Düşürme:** Daha az özellikle çalışan algoritmalar çok daha hızlıdır.

**PCA Algoritması Nasıl Çalışır?**

1. **Veriyi Standartlaştırma (Standardization):**

   * **Neden?** PCA, varyans maksimizasyonu üzerine kuruludur. **Eğer özelliklerimiz farklı ölçeklerdeyse** (örneğin, maaş[10,000-100,000], yaş[20-60]), maaş özelliği varyansı domine eder ve PCA sonucunu yanıltır.
   * **Nasıl?** Her bir özellik sütunundan kendi ortalamasını çıkarıp standart sapmasına bölerek tüm özellikleri ortalaması 0, standart sapması 1 olan bir ölçeğe getiririz.

$$
X_{std} = \frac {X - \mu} {\sigma}
$$

* $X$ : Bu, veri setindeki belirli bir **veri noktası** (ya da örnek) anlamına gelir.
* $\mu$ : Veri setinin **ortalamasıdır.**
* $\sigma$ : Veri setinin **standart sapmasıdır**
* $X_{std}$ : **standardize edilmiş** veri değerini temsil eder. 

2. **Kovaryans Matrisini Hesaplama:**
   * Verinin değişkenleri arasındaki ilişkiyi gösteren kovaryans matrisi hesaplanır. En yüksek varyans yönlerini bulmaktır.

$$
Cov(X,Y) = \frac {1} {n-1} \sum _{i=1}^{n} (X_i - \hat X) (Y_i - \hat Y)
$$

* $X_i$ ve $Y_i$ : Her bir i, X ve Y veri setindeki belirli bir gözlemi temsil eder.
* $\hat X$ ve $\hat Y$ : X ve Y veri setlerinin **ortalamasıdır**
* $n$ : Veri setindeki gözlem sayısı

3. **Özdeğer ve Özvektörleri Bulma**
   * Kovaryans matrisinin **özvektörleri** (principal components) ve özdeğerleri hesaplanır.
   * Özdeğerler, her bir bileşenin ne kadar varyans taşıdığını gösterir.

4. **Özdeğerlere Göre Sıralama ve Bileşen Seçimi**
   * Özdeğerler büyükten küçüğe sıralanır.
   * En yüksek özdeğere sahip özvektörler seçilir (genellikle toplam varyansın ~%95'ini koruyacak şekilde).

5. **Yeni Alt Uzayı Oluşturma (Dönüşüm)**
   * Son adım, orijinal standartlaştırılmış veriyi, seçtiğimiz temel bileşenlerin (özvektörlerin) oluşturduğu yeni uzaya "projelendirmektir" (dönüştürmektir).
   * Bu, orijinal veri matrisini, seçilen özvektörlerden oluşan matrisle çarparak yapılır (dot product).
   * Sonuçta, orijinal N x D boyutlu veri seti, N x K boyutlu yeni bir veri setine dönüşür (burada K < D).

**Avantajları:**
* **Hesaplama verimliliği:** Yüksek boyutlu verilerde hız sağlar.
* **Gürültü filtresi:** Düşük varyanslı bileşenler genellikle gürültüdür.
* **Görselleştirme:** Veriyi 2D veya 3D'ye indirgeyerek anlaşılır hale getirir.

**Dezavantajları:**
* **Yorumlanabilirlik:** Yeni bileşenler orijinal özelliklerle doğrudan ilişkili değildir.
* **Doğrusallık varsayımı:** PCA, doğrusal ilişkileri temel alır. Doğrusal olmayan veriler için Kernel PCA veya t-SNE gibi yöntemler daha uygundur.
* **Varyans kaybı:** Boyut azaltma sırasında bilgi kaybı olabilir.

**Kullanım Alanları**
* **Yüz tanıma**,
* **Biyoinformatik** (Gen ekspresyon analizi),
* **Finans** (Risk faktörlerinin analizi),
* **Doğal** dil işleme (Kelime vektörlerinin sıkıştırılması).


```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

# Örnek veri
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])

# Standartlaştırma
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# PCA uygulama (1 boyuta indirgeme)
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_std)

print("Orijinal boyut:", X.shape)
print("İndirgenmiş boyut:", X_pca.shape)
print("Açıklanan varyans oranı:", pca.explained_variance_ratio_) # %100 varyans korundu (basit örnek olduğu için)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Iris veri setini yükleyelim
iris = load_iris()

# Veriyi daha kolay kullanmak için bir pandas DataFrame'e dönüştürelim
# iris.data -> özellikler (4 sütun)
# iris.target -> etiketler (çiçek türleri: 0, 1, 2)
# iris.feature_names -> özelliklerin isimleri
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print("Veri setinin ilk 5 satırı:")
print(df.head())
print("\nVeri setindeki özellik sayısı:", len(iris.feature_names))
```

```python
# PCA, varyansa duyarlı olduğu için, farklı ölçeklerdeki özellikleri aynı kefeye koymamız gerekiyor. Bunun için StandardScaler kullanacağız. Bu işlem, her özelliğin ortalamasını 0 ve standart sapmasını 1 yapacaktır.

# Sadece özellik sütunlarını seçelim
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
x = df.loc[:, features].values

# Veriyi standartlaştıralım
x_scaled = StandardScaler().fit_transform(x)

# Standartlaştırılmış veriyi bir DataFrame'e dönüştürerek görelim
scaled_df = pd.DataFrame(data=x_scaled, columns=features)
print("\nStandartlaştırılmış verinin ilk 5 satırı:")
print(scaled_df.head())
```

```python
# Şimdi scikit-learn'in PCA sınıfını kullanarak boyut azaltma işlemini gerçekleştireceğiz. Amacımız 4 boyuttan 2 boyuta inmek olduğu için n_components=2 parametresini kullanacağız.


# PCA modelini oluşturalım (2 bileşen istiyoruz)
pca = PCA(n_components=2)

# Modeli standartlaştırılmış veriye uygulayalım
principal_components = pca.fit_transform(x_scaled)

# Yeni oluşturulan temel bileşenleri bir DataFrame'e dönüştürelim
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

print("\nPCA sonrası oluşturulan 2 boyutlu verinin ilk 5 satırı:")
print(pca_df.head())

# Bu yeni PCA verisini orijinal DataFrame'in yanına ekleyelim
final_df = pd.concat([pca_df, df[['species']]], axis=1)
print("\nSon birleştirilmiş DataFrame:")
print(final_df.head())
```

```python
# PCA sadece veriyi dönüştürmekle kalmaz, aynı zamanda bize bu yeni bileşenlerin ne kadar "önemli" olduğunu da söyler. explained_variance_ratio_ özelliği, her bir temel bileşenin orijinal verideki toplam varyansın ne kadarını açıkladığını gösterir.

# Her bir temel bileşenin açıkladığı varyans oranı
explained_variance = pca.explained_variance_ratio_

print(f"\nBirinci Temel Bileşen (PC1) Varyans Oranı: {explained_variance[0]:.2f}")
print(f"İkinci Temel Bileşen (PC2) Varyans Oranı: {explained_variance[1]:.2f}")
print(f"Toplam Açıklanan Varyans (PC1 + PC2): {np.sum(explained_variance):.2f}")
```

* **PC1, toplam varyansın yaklaşık %73'ünü** tek başına açıklıyor. Bu, verideki en önemli değişim yönünün bu bileşen olduğunu gösterir.
* **PC2, geriye kalan varyansın yaklaşık %23'ünü** açıklıyor.

### 2.2 t-SNE (t-Distributed Stochastic Neighbor Embedding) (t-Dağılımlı Stokastik Komşu Gömme):

t-SNE'nin temel amacı, **yüksek boyutlu uzaydaki noktalar arasındaki komşuluk ilişkilerini (benzerliklerini), düşük boyutlu bir uzayda (genellikle 2D veya 3D) olabildiğince korumaktır.** PCA'dan farklı olarka, doğrusal olmayan (non-linear) yapıları yakalamada çok daha başarılıdır.

* Yüksek boyutlu uzayda **birbirine yakın olan noktalar,** düşük boyutlu haritalarda da **birbirine yakın olmalıdır.**
* Yüksek boyutlu uzayda **birbirinden uzak olan noktalar,** düşük boyutlu haritada da **birbirinden uzak olmalıdır.**

PCA gibi global varyansı korumaya odaklanmak yerine, t-SNE **yerel (local) komşuluk yapılarını** korumaya odaklanır. Bu yüzden karmaşık,  kıvrımlı ve içe içe geçmiş veri yapılarının (manifoldların) görselleştirilmesinde çok etkili olduğu için kullanırız.

Örnek: Bir sosyal ağdaki binlerce insan arasındaki arkadaşlık ilişkilerini düşün. Bu ilişkiler çok boyutlu olabilir (ortak arkadaşlar, ortak okullar, ortak iş yerler, ortak ilgi alanları vb.)

* **t-SNE'ni Görevi:** Bu karmaşık ilişki ağını alıp, bir kağıt üzerinde (2D) bir haritaya dönüştürmek.
* **Sonuç:** Bu haritada, sıkı arkadaş grupları bir arada kümelenmiş olarak görünürken, birbiriyle pek ilgisi olmayan insanlar haritanın farklı köşelerinde yer alacaktır. t-SNE, bu "grupları" ve "toplulukları" görsel olarak ortaya çıkarır.

**Algoritma**

**Adım 1: Yüksek Boyutlu Uzayda Olasılıksal Benzerlikleri Hesaplama**

* Her bir veri noktası $x_i$ için, diğer noktalara olan benzerlik **koşullu olasılık** $p_{j|i}$ ile ifade edilir.
* $x_j$'nin $x_i$'nin komşusu olma olasılığı denklemi:
    $$
    p_{j|i} = \frac{
        \exp\left(-\frac{\| x_i - x_j \|^2}{2\sigma_i^2}\right)
    }{
        \sum_{k \neq i} \exp\left(-\frac{\| x_i - x_k \|^2}{2\sigma_i^2}\right)
    }
    $$

  * $\sigma_i, \space x_i$ noktasının etrafındaki dağılımın genişliğini kontrol eder (perplexity parametresiyle ayarlanır)
* **Simetrikleştirme:** $p_{ij} = \frac {p_{j|i}+ p_{i|j}} {2N}$ şeklinde ortak olasılık dağılımı oluşturulur



* t-SNE, noktalar arasındaki Öklid mesafesini direkt kullanmak yerine, bu mesafeleri **olasılıksal bir benzerlik dağılımına** dönüştürür.
* Her bir veri noktası i için, diğer tüm noktaların j onun komşusu olma olasılığını $p_{j|i}$ hesaplar.
* Bu olasılık, i noktası merkezli bir **Gauss (Normal) Dağılımı** kullanılarak hesaplanır.
  * i noktasına yakın olan j noktaları için bu olasılık değeri **yüksek** olur.
  * i noktasına uzak olan j noktaları için bu olasılık değeri **düşük** olur.
* **Perplexity (Şaşkınlık) Parametresi:** Bu adımda önemli bir parametre vardır: perplexity. Bu parametre, her bir noktanın ne kadar komşuyu dikkate alacağını belirler. Düşük perplexity değeri daha yerel yapılara odaklanırken, yüksek değer daha global yapılara odaklanır. scikit-learn'de genellikle 5 ile 50 arasında bir değer seçilir.


<br></br>

**Adım 2: Düşük Boyutlu Uzayda Başlangıç Haritası Oluşturma**

* Veri noktaları, düşük boyutlu uzayda (örneğin 2D) rastgele konumlara yerleştirilir. Bu, henüz hiçbir anlamı olmayan bir başlangıç haritasıdır.
$$
q_{ij} = \frac{
    (1 + \| y_i - y_j \|^2)^{-1}
}{
    \sum_{k \neq l} (1 + \| y_k - y_l \|^2)^{-1}
}
$$


<br></br>

**Adım 3: İki Dağılım Arasındaki Farkın Minimizasyonu**

* t-SNE, yüksek boyutlu $p_{ij}$ ve düşük boyutlu $q_{ij}$ dağılımlar arasındaki farkı **Kullback-Leibler (KL) Diverjansı** ile ölçer.

    $$
    KL(P||Q) = \sum _{i \neq j} p_{ij}log {\frac {p_{ij}} {q_{ij}}}
    $$

* Optimizasyon genellikle **gradient descent** ile yapılır.

**Önemli Noktalar**: t-SNE çok güçlü olduğu kadar yanlış yorumlanmaya çok açık bir algoritmadır.

1. **Küme Boyutları Anlamlı Değildir:** Bir kümenin haritada kapladığı alan, o kümenin içindeki varyans veya nokta sayısı hakkında bilgi vermez. t-SNE, yoğun kümeleri de seyrek kümeleri de genişleterek benzer boyutlarda gösterebilir.
2. **Kümeler Arası Mesafe Anlamlı Değildir:** Haritada iki küme arasındaki boşluğun büyüklüğü, bu kümelerin gerçekte ne kadar farklı olduğu anlamına gelmez. Sadece "farklı" olduklarını söyler. Birbirine yakın görünen iki ayrı küme, çok uzak görünen iki ayrı kümeden daha benzer olmayabilir.
3. **Parametrelere Duyarlılık:** Perplexity ve n_iter (iterasyon sayısı) gibi parametreler, sonuç haritasını önemli ölçüde değiştirebilir. En iyi sonucu bulmak için farklı parametrelerle denemeler yapmak gerekir.
4. **Sadece Görselleştirme İçin Kullanılmalı:** t-SNE, bir makine öğrenmesi modeline girdi olarak kullanılacak özellikler oluşturmak için genellikle iyi bir seçim değildir. Bu iş için PCA veya Autoencoder'lar uygundur. t-SNE'nin asıl amacı, insanların veriyi anlaması için bir harita çizmektir.

* **Bilgisayarlı Görü:** Derin öğrenme modellerinin (CNN) görüntülerden öğrendiği özellik uzayını görselleştirmek.
* **Doğal Dil İşleme (NLP):** Kelime gömme (word embeddings) modellerinin (Word2Vec, GloVe) sonuçlarını görselleştirerek kelimeler arasındaki anlamsal ilişkileri görmek. (kelimelerin konumları)
* **Biyoinformatik:** Gen ekspresyon verilerini (RNA-Seq) veya tek hücre verilerini (single-cell) analiz ederek hücre türlerini ve alt gruplarını keşfetmek.
* **Siber Güvenlik:** Ağ trafiği verilerindeki normal ve anormal davranış kalıplarını görsel olarak ayırt etmek.


```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# Örnek veri: MNIST rakamları
digits = load_digits()
X = digits.data
y = digits.target

# t-SNE uygula
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

# Görselleştirme
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', alpha=0.6)
plt.colorbar()
plt.title("t-SNE ile MNIST Görselleştirme")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml

# MNIST veri setini scikit-learn'den yükleyelim
# Not: Bu işlem internet hızınıza bağlı olarak birkaç dakika sürebilir.
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='liac-arff')

# Veri setinden özellikler (X) ve etiketleri (y) alalım
X = mnist.data
y = mnist.target.astype(int) # Etiketleri integer'a çevirelim

print("Orijinal Veri Setinin Boyutu:", X.shape)
print("Etiketlerin Boyutu:", y.shape)

# Hesaplama süresini kısaltmak için veri setinden rastgele 5000 örnek alalım
np.random.seed(42) # Sonuçların tekrarlanabilir olması için
sample_indices = np.random.choice(X.shape[0], 5000, replace=False)
X_sample = X[sample_indices]
y_sample = y[sample_indices]

print("\nKullanılacak Örnek Veri Setinin Boyutu:", X_sample.shape)
```

```python
# Rastgele birkaç rakamı çizdirelim
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    # 784 boyutlu vektörü 28x28'lik resme geri dönüştürelim
    plt.imshow(X_sample[i].reshape(28, 28), cmap='gray_r')
    plt.title(f'Etiket: {y_sample[i]}')
    plt.axis('off')
plt.suptitle('MNIST Veri Setinden Örnek Rakamlar', fontsize=16)
plt.show()
```

```python
print("t-SNE başlıyor... Bu işlem birkaç dakika sürebilir.")

# t-SNE modelini oluşturalım
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)

# Modeli 784 boyutlu veriye uygulayalım
tsne_results = tsne.fit_transform(X_sample)

print("t-SNE tamamlandı.")
print("Yeni 2 Boyutlu Verinin Boyutu:", tsne_results.shape)

# Sonuçları daha kolay kullanmak için bir DataFrame'e dönüştürelim
tsne_df = pd.DataFrame(data=tsne_results, columns=['tsne-2d-one', 'tsne-2d-two'])
tsne_df['y'] = y_sample # Orijinal etiketleri ekleyelim
```

```python
plt.figure(figsize=(14, 10))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="y", # Noktaları rakam etiketlerine göre renklendir
    palette=sns.color_palette("hls", 10), # 10 farklı renk kullan
    data=tsne_df,
    legend="full",
    alpha=0.8 # Noktaların şeffaflığı
)

plt.title('MNIST Rakamlarının t-SNE ile 2 Boyutta Görselleştirilmesi', fontsize=18)
plt.xlabel('t-SNE Boyut 1', fontsize=14)
plt.ylabel('t-SNE Boyut 2', fontsize=14)
plt.legend(title='Rakamlar')
plt.show()
```

```python
# K-Means örnek


import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Tekrarlanabilir sonuçlar için random state belirleyelim
np.random.seed(42)

# 1. Sıradan Oyuncular (Casuals): Az saat, az harcama
# 100 oyuncu, ortalama 5 saat oyun, 25 TL harcama
casual_oyuncular = np.random.randn(100, 2) * np.array([3, 10]) + np.array([5, 25])

# 2. Adanmış Oyuncular (Grinders): Çok saat, az harcama
# 100 oyuncu, ortalama 30 saat oyun, 50 TL harcama
adanmis_oyuncular = np.random.randn(100, 2) * np.array([5, 20]) + np.array([30, 50])

# 3. Balinalar (Whales): Orta/Çok saat, çok harcama
# 50 oyuncu, ortalama 20 saat oyun, 300 TL harcama
balina_oyuncular = np.random.randn(50, 2) * np.array([8, 50]) + np.array([20, 300])

# Tüm oyuncu verilerini birleştirelim
oyuncular = np.concatenate([casual_oyuncular, adanmis_oyuncular, balina_oyuncular])

# Veri setini karıştıralım ki sıralı olmasın
np.random.shuffle(oyuncular)

print(f"Toplam oyuncu sayısı: {len(oyuncular)}")
print("İlk 5 oyuncunun verisi (Oyun Saati, Harcama):")
print(oyuncular[:5])

plt.figure(figsize=(10, 7))
# oyuncular[:, 0] -> x ekseni (oyun saati)
# oyuncular[:, 1] -> y ekseni (harcama)
plt.scatter(oyuncular[:, 0], oyuncular[:, 1], s=50, alpha=0.6, edgecolors='k')
plt.title('Oyuncu Davranışlarının Ham Dağılımı (Etiketsiz)')
plt.xlabel('Haftalık Ortalama Oyun Süresi (Saat)')
plt.ylabel('Toplam Harcama (TL)')
plt.grid(True)
plt.show()


# K-Means modelini oluşturalım
# n_clusters: Küme sayısı (K)
# n_init: Farklı merkez başlangıçlarıyla algoritmanın kaç kez çalıştırılacağı. En iyi sonuç seçilir.
# random_state: Sonuçların tekrarlanabilir olmasını sağlar.
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)

# Modeli verilerimizle eğitelim
kmeans.fit(oyuncular)

# Her bir oyuncunun hangi kümeye atandığını gösteren etiketleri alalım
kumeler = kmeans.labels_

# Algoritmanın bulduğu küme merkezlerinin (centroid) koordinatlarını alalım
merkezler = kmeans.cluster_centers_

print("Her oyuncu için atanan küme numaraları (ilk 10 oyuncu):")
print(kumeler[:10])
print("\nBulunan küme merkezleri (centroid'ler):")
print("Merkez   | Oyun Saati (Ort.) | Harcama (Ort.)")
print("---------------------------------------------")
for i, merkez in enumerate(merkezler):
    print(f"Küme {i}    | {merkez[0]:.2f}             | {merkez[1]:.2f}")


    plt.figure(figsize=(12, 8))

# K-Means tarafından renklendirilmiş oyuncu verilerini çizdirelim
# c=kumeler argümanı, her noktayı kendi küme numarasına göre renklendirir.
scatter = plt.scatter(oyuncular[:, 0], oyuncular[:, 1], c=kumeler, s=50, cmap='viridis', alpha=0.7, edgecolors='k')

# Küme merkezlerini (centroid'leri) büyük kırmızı 'X' işaretiyle çizdirelim
plt.scatter(merkezler[:, 0], merkezler[:, 1], c='red', s=250, marker='X', label='Küme Merkezleri (Centroids)')

plt.title('K-Means ile Oluşturulan Oyuncu Segmentleri')
plt.xlabel('Haftalık Ortalama Oyun Süresi (Saat)')
plt.ylabel('Toplam Harcama (TL)')
plt.legend()
plt.grid(True)
plt.show()
```
